```python
  ** Different Types of Data visulization 
    
   >> univariate Analysis: in the univariate analysis, we will be using a single feature 
        to analyze almost the properties.
   >>  Bivariate Analysis : in this when we comparethe data b/w exactly 2 features then it is 
        known as bivariate analysis.
   >> multivaite Analysis: in the multivariate analysis, we will be comparing more than 2 variables.
    
    Matplotlib
    
    Seaborn
    
    Plotly
    
    i) Scatter plot(B)
     
    ii) Pair plot(M)
    
    iii) Box plot(U)
    
    iv) Violin plot(U)
    
    v) Distribution plot(U)
    
    vi) Joint plot(u) & (B)
    
    vii) Bar chart(B)
    
    viii)Line plot(B)
```

```python
# initilazing the data 

import matplotlib.pyplot as plt

x = [10, 20, 30, 40]
y = [30, 22, 45, 50 ]

# plotting the data

plt.plot(x, y)

plt.show()
```

```python
# initalizing the data 

x = [10, 20, 30, 40]
y = [30, 22, 45, 50]

# plotting the data

plt.plot(x,y)

# Adding label on the y-axis

plt.ylabel('y-axis')

# Adding lebel on x- axis

plt.xlabel('x-axis')

# setting the label of x-axis

plt.xticks(x, labels=["one", "two", "three", "four"])

# Adding Legends

plt.legend(["Linear Graph"])

plt.show()
```
